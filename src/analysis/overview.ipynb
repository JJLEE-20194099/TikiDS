```python
merge_rating_item = pd.merge(rating, items, how='inner', left_on = 'item id', right_on='id')
print(len(merge_rating_item))
merge_rating_item.head()
```

```python
num_rating_per_category = merge_rating_item.groupby('category').count()['rating']
print(num_rating_per_category)
average_rating_per_category = merge_rating_item.groupby('category').mean()['rating']
print(average_rating_per_category)
```

```python
rating_data = pd.read_csv('../../data/datasets/ratings/data.csv', sep=',', encoding='utf-8')
rating_data = rating_data[rating_data.groupby('user id')['user id'].transform('count').ge(5)]

formated_user_ids = rating_data['user id'].value_counts().index.tolist()
formated_item_ids = rating_data['item id'].value_counts().index.tolist()


rating_data['formated user id'] = [0] * len(rating_data)
rating_data['formated item id'] = [0] * len(rating_data)

rating_data_arr = rating_data.values


for i in range(len(formated_user_ids)):
    rows = np.where(rating_data_arr[:, 0] == formated_user_ids[i])
    rating_data_arr[rows, 5] = i

for i in range(len(formated_item_ids)):
    rows = np.where(rating_data_arr[:, 1] == formated_item_ids[i])
    rating_data_arr[rows, 6] = i
```
